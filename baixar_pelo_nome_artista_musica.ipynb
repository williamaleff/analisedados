```python
import yt_dlp
import os

# === CONFIGURAÇÕES ===
# Pasta de destino
output_path = r"D:\william\musicas em mp3"

# Lista com os nomes das músicas (copiados do Spotify)
# Dica: use "Artista - Nome da Música"
search_queries = [
    "Ivonaldo Albuquerque - Sua Presença",
    "Canção e louvor - Comum e imperfeito",
]

# Cria a pasta, se não existir
os.makedirs(output_path, exist_ok=True)

# === CONFIGURAÇÃO DO yt-dlp ===
ydl_opts = {
    "format": "bestaudio/best",
    "outtmpl": os.path.join(output_path, "%(title)s.%(ext)s"),
    "postprocessors": [{
        "key": "FFmpegExtractAudio",
        "preferredcodec": "mp3",
        "preferredquality": "192",
    }],
    "noplaylist": True,
    "quiet": False,  # mostre o progresso
    # Se o FFmpeg não estiver no PATH, defina o caminho manualmente:
    # "ffmpeg_location": r"C:\ffmpeg\bin",
}

# === LOOP DE DOWNLOAD ===
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    for query in search_queries:
        print(f"\n🔍 Procurando: {query}")
        try:
            # "ytsearch1:" procura no YouTube e pega o primeiro resultado
            info = ydl.extract_info(f"ytsearch1:{query}", download=True)
            if "entries" in info:
                entry = info["entries"][0]
                print(f"✅ Baixado: {entry['title']}")
            else:
                print("⚠️ Nenhum resultado encontrado.")
        except Exception as e:
            print(f"❌ Erro ao processar '{query}': {e}")

print("\n🎉 Todos os downloads e conversões foram concluídos!")
```
