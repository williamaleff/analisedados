```python
# 📌 1. Importações
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from time import sleep
from IPython.display import display
from selenium.webdriver.chrome.options import Options

# 📌 2. Lê a coluna A da planilha "custodiados"
arquivo_excel = r"D:/custodiados"
nome_arquivo = "custodiados.xlsx"
caminho_completo = r"D:\custodiados.xlsx"

df = pd.read_excel(caminho_completo, sheet_name="custodiados", usecols="A")
custodiados = df.iloc[:, 0].dropna().astype(str).tolist()

# 📌 3. Inicia o navegador
chrome_options = Options()
chrome_options.add_argument("--start-maximized")

servico = Service('D:/chromedriver.exe')  # coloque o caminho completo se necessário
driver = webdriver.Chrome(service=servico, options=chrome_options)

# 📌 4. Acessa o sistema e faz login
driver.get("http://172.18.4.44:3000/painelSistema")
sleep(2)

# Preenche login e senha
driver.find_element(By.XPATH, "/html/body/div/div/div/div/div/div/div[2]/form/div[1]/div/div/div/div/span/span/input").send_keys("elisangela.helcias")
driver.find_element(By.XPATH, "/html/body/div/div/div/div/div/div/div[2]/form/div[2]/div/div/div/div/span/span/input").send_keys("30069218")

# Clica no botão de login
driver.find_element(By.XPATH, "/html/body/div/div/div/div/div/div/div[2]/form/div[4]/div/button").click()

# Aguarda a próxima tela
sleep(3)
driver.find_element(By.XPATH, "/html/body/div/div/div/div/div/div[2]/div/div[1]").click()

# Aguarda o carregamento
sleep(15)

# 📌 5. Loop pelos custodiados
resultados = []

for idx, codigo in enumerate(custodiados, 1):
    url = f"http://172.18.4.44:3000/ficha/custodiado/{codigo}"
    driver.get(url)
    sleep(5)
    
    try:
        xpath_data = "/html/body/div/section/section/main/div/div/div/div[2]/div/div/div/div[1]/div/div/div[2]/table/tbody/tr[5]/td[3]/span[2]"
        elemento = driver.find_element(By.XPATH, xpath_data)
        data_valor = elemento.text.strip()
    except Exception as e:
        data_valor = "NÃO ENCONTRADO"
    
     # Armazena o resultado
    resultado = (idx, codigo, data_valor)
    resultados.append(resultado)

    # 🟢 Exibe imediatamente o resultado
    print(f"{resultado[0]}, {resultado[1]}, {resultado[2]}")

# 📌 6. Exibir resultados no final
for linha in resultados:
    print(f"{linha[0]}, {linha[1]}, {linha[2]}")

# 📌 7. Fecha o navegador
driver.quit()
```
