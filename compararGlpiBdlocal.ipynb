```python
import pandas as pd

# Carregar os dois arquivos CSV
df_chamado_glpi = pd.read_csv('D:/suporte08092025.csv')  # Arquivo com chamados
df_id = pd.read_csv('D:/glpi.csv')  # Arquivo com IDs

# Verificar o nome das colunas
print(df_chamado_glpi.columns)
print(df_id.columns)

# Suponhamos que a coluna que contém os IDs em ambos os arquivos se chama "ID"
# Caso o nome da coluna seja diferente, altere "ID" para o nome correto
df_chamado_glpi['chamado_glpi'] = df_chamado_glpi['chamado_glpi'].fillna(0).astype(int).astype(str)

# Converter as colunas de ID para listas
ids_chamado_glpi = df_chamado_glpi['chamado_glpi'].apply(lambda x: x.replace(" ", "")).tolist()  # Remove todos os espaços
ids_id = df_id['ID'].apply(lambda x: x.replace(" ", "")).tolist()  # Remove todos os espaços

# Exibir os valores de 'ID' após a conversão
print("\nValores de 'chamado_glpi' após a conversão:")
print(ids_chamado_glpi)
print(ids_id)
# Identificar IDs da lista 'ID' que não estão na lista 'chamado_glpi'
ids_nao_encontrados = [id for id in ids_id if id not in ids_chamado_glpi]

# Exibir ou salvar os IDs que não estão na lista 'chamado_glpi'
print("IDs não encontrados em chamado_glpi:", ids_nao_encontrados)

# Caso queira salvar esses IDs em um novo arquivo CSV
df_nao_encontrados = pd.DataFrame(ids_nao_encontrados, columns=["ID"])
df_nao_encontrados.to_csv("D:/ids_nao_encontrados5.csv", index=False)
```
