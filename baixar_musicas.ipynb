```python
import os
import yt_dlp

# Pasta de destino
output_path = r"D:\william\musicas em mp3"

# Lista de links do YouTube
urls = [
    "https://www.youtube.com/watch?v=0mJSNgnl8y8",
    "https://www.youtube.com/watch?v=6OnuqNFeGUY",
    "https://www.youtube.com/watch?v=8AESIsViPsg",
    "https://www.youtube.com/watch?v=BzYPyhE-Acg",
    "https://www.youtube.com/watch?v=jr1qyQQw2NI",
    "https://www.youtube.com/watch?v=aW5a2oewZpI",
    "https://www.youtube.com/watch?v=dNMIxWWxYNo",
    "https://www.youtube.com/watch?v=Ds200TtR97o",
    "https://www.youtube.com/watch?v=r3QwS4eH7KI",
    "https://www.youtube.com/watch?v=8BLpVdFvDfE",
    "https://www.youtube.com/watch?v=700DisbyEOo",
]

# Cria a pasta se não existir
os.makedirs(output_path, exist_ok=True)

# Configuração do yt-dlp
ydl_opts = {
    "format": "bestaudio/best",
    "outtmpl": os.path.join(output_path, "%(title)s.%(ext)s"),
    "postprocessors": [{
        "key": "FFmpegExtractAudio",
        "preferredcodec": "mp3",
        "preferredquality": "192",
    }],
    "noplaylist": True,
    "quiet": False,
}

# Faz o download de todas as músicas
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download(urls)

print("✅ Todos os downloads foram concluídos com sucesso!")
```
