```python
from pypdf import PdfWriter
import os

# === CONFIGURAÇÕES ===
pasta_pdfs = r"D:\william\documentos"
arquivo_saida = r"D:\william\documentos\PDF_UNIFICADO.pdf"

# Pega todos os PDFs da pasta, em ordem alfabética
pdfs = [f for f in os.listdir(pasta_pdfs) if f.lower().endswith(".pdf")]
pdfs.sort()

writer = PdfWriter()

for pdf in pdfs:
    caminho = os.path.join(pasta_pdfs, pdf)
    print(f"📄 Adicionando: {pdf}")
    writer.append(caminho)

# Grava o arquivo final
with open(arquivo_saida, "wb") as f_out:
    writer.write(f_out)

print(f"\n✅ PDF final criado com sucesso: {arquivo_saida}")
```
